```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 500


customers = pd.DataFrame({
    "customer_id": range(1, n+1),
    "customer_name": [f"Customer_{i}" for i in range(1, n+1)],
    "country": np.random.choice(
        ["India","USA","UK"],
        n
    ),
    "signup_date": pd.date_range(
        start="2024-01-01",
        periods=n,
        freq="D"
    )
})


subscriptions = pd.DataFrame({
    "customer_id": range(1,n+1),
    "plan": np.random.choice(
        ["Basic","Premium","Enterprise"],
        n
    ),
    "monthly_fee": np.random.choice(
        [499,999,1999],
        n
    ),
    "status": np.random.choice(
        ["Active","Expired"],
        n,
        p=[0.85,0.15]
    )
})


product_usage = pd.DataFrame({
    "customer_id": range(1,n+1),
    "last_login_days": np.random.randint(
        1,120,n
    ),
    "monthly_logins": np.random.randint(
        0,50,n
    ),
    "feature_usage": np.random.randint(
        1,100,n
    )
})


support_tickets = pd.DataFrame({
    "customer_id": range(1,n+1),
    "total_tickets": np.random.randint(
        0,10,n
    ),
    "unresolved_tickets": np.random.randint(
        0,5,n
    )
})


payments = pd.DataFrame({
    "customer_id": range(1,n+1),
    "failed_payments": np.random.randint(
        0,4,n
    )
})


interventions = pd.DataFrame({
    "customer_id": range(1,n+1),
    "owner": np.random.choice(
        ["Retention Team A",
         "Retention Team B"],
        n
    ),
    "action_taken": np.random.choice(
        ["Call",
         "Discount",
         "Email"],
        n
    ),
    "intervention_status": np.random.choice(
        ["Pending",
         "Contacted",
         "Retained",
         "Churned"],
        n
    )
})

customers.to_csv("customers.csv",index=False)
subscriptions.to_csv("subscriptions.csv",index=False)
product_usage.to_csv("product_usage.csv",index=False)
support_tickets.to_csv("support_tickets.csv",index=False)
payments.to_csv("payments.csv",index=False)
interventions.to_csv("interventions.csv",index=False)

print("All CSV files created successfully!")
```
